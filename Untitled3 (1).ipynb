```python
# data_loader.py

import pandas as pd
import cv2
from pathlib import Path

def load_data(base_path):
    
    data_dict = {}
    base_path = Path(base_path)
    for folder in base_path.iterdir():
        if folder.is_dir():
            csv_file = next(folder.glob("*.csv"), None)
            video_file = next(folder.glob("*.mov"), None) or next(folder.glob("*.mp4"), None)
            
         
            if csv_file:
                try:
                    temperature_data = pd.read_csv(csv_file, on_bad_lines='skip')  
                    print(f"Loaded temperature data from {csv_file.name}")
                except pd.errors.ParserError:
                    print(f"Error reading {csv_file.name}. Skipping this file.")
                    temperature_data = None
            else:
                temperature_data = None
            
            # Load video file
            video_reader = None
            if video_file:
                video_reader = cv2.VideoCapture(video_file.as_posix())
                if not video_reader.isOpened():
                    print(f"Error: Could not open video {video_file.name}")
                    video_reader = None
                else:
                    print(f"Loaded video from {video_file.name}")

            data_dict[folder.name] = {
                "temperature_data": temperature_data,
                "video_file": video_reader
            }
    
    return data_dict
```

```python
# frame_processor.py
import os
import pandas as pd
import cv2
import numpy as np
from pathlib import Path
from ultralytics import YOLO
import csv
# Load YOLO model
yolo_model = YOLO("best.pt")  # Use the small YOLOv8 model for efficiency

def data_load(base_path):
    """Load temperature data from CSV and video frames from video files."""
    data_dict = {}
    base_path = Path(base_path)
    
    for folder in base_path.iterdir():
        if folder.is_dir():
            csv_file = next(folder.glob("*.csv"), None)
            video_file = next(folder.glob("*.mov"), None) or next(folder.glob("*.mp4"), None)
            
            # Load temperature data
            if csv_file:
                try:
                    temperature_data = pd.read_csv(csv_file, on_bad_lines='skip')  # Skip bad rows
                    print(f"Loaded temperature data from {csv_file.name}")
                except pd.errors.ParserError as e:
                    print(f"Error reading {csv_file.name}: {e}")
                    temperature_data = None
            else:
                temperature_data = None
            
            # Load video file
            video_reader = None  # Initialize as None
            if video_file:
                video_reader = cv2.VideoCapture(video_file.as_posix())
                if not video_reader.isOpened():
                    print(f"Error: Could not open video {video_file.name}")
                    video_reader = None
                else:
                    print(f"Loaded video from {video_file.name}")

            # Store data in dictionary
            data_dict[folder.name] = {
                "temperature_data": temperature_data,
                "video_file": video_reader
            }
    
    return data_dict


def detect_fire_yolo(frame, yolo_model, confidence_threshold=0.3):
    """Detect fire in a frame using YOLOv8 model."""
    results = yolo_model(frame)  # Get YOLO model results on the frame
    
    fire_detected = False
    fire_area = 0
    
    # Iterate over detections in `results.boxes` if available
    if results and results[0].boxes is not None:
        for box in results[0].boxes:
            # Extract confidence, bounding box coordinates, and class ID
            conf = box.conf[0].item()  # Confidence score
            cls_id = box.cls[0].item()  # Class ID
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
            
            if conf > confidence_threshold and cls_id == 0:  # Assuming '0' is fire class
                fire_detected = True
                fire_area += (x2 - x1) * (y2 - y1)
                
                # Draw bounding box around detected fire
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    return fire_detected, fire_area

def process_frames(data_dict, temp_threshold=60, area_threshold=5000, output_file="fire_detection_results.csv"):
    """Process video frames and check if fire is detected and if it is potentially harmful."""
    results = []  
    
    for folder_name, data in data_dict.items():
        temperature_data = data["temperature_data"]
        video_reader = data["video_file"]
        
        if temperature_data is not None and video_reader is not None:
            success, frame = video_reader.read()
            frame_index = 0
            while success:
                if frame_index < len(temperature_data):
                    temp_row = temperature_data.iloc[frame_index]
                    frame_temperature = temp_row.get("Temperature", 0)
                    
                   
                    confidence_threshold = 0.3 if frame_temperature > temp_threshold else 0.5
                    
                    fire_detected, fire_area = detect_fire_yolo(frame, yolo_model, confidence_threshold)
                    
                    severity = "Harmful" if fire_detected and fire_area > area_threshold else "Non-harmful"
                    results.append([folder_name, frame_index, frame_temperature, fire_area, severity])
                    
                    if fire_detected and fire_area > area_threshold:
                        print(f"Warning: Harmful fire detected in {folder_name} - Frame {frame_index}")
                    else:
                        print(f"Fire detected but not harmful in {folder_name} - Frame {frame_index}")
                
                success, frame = video_reader.read()
                frame_index += 1
            video_reader.release()
    
    # Save results to a CSV file
    save_results(results, output_file)

def save_results(results, output_file="fire_detection_results.csv"):
    """Save detection results to a CSV file."""
    with open(output_file, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Folder Name", "Frame Index", "Temperature", "Fire Area", "Severity"])
        for result in results:
            writer.writerow(result)
if __name__ == "__main__":
    base_path = r'C:\Users\rekha\OneDrive\Desktop\archive\Fire'
    data_dict = data_load(base_path)
    process_frames(data_dict)
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# Initialize YOLO model
yolo_model = YOLO("best.pt")

# Ground truth labels (1 for fire, 0 for no fire)
# Replace with actual labels for your dataset (frame-wise labels)
ground_truth = [1, 0, 1, 1, 0, 0, 1, 0, 0, 1]  # Example labels for testing
predictions = []  # To store model predictions

# Fire detection function
def detect_fire_yolo(frame, yolo_model, confidence_threshold=0.5):
    results = yolo_model(frame)
    fire_detected = False

    if results and results[0].boxes is not None:
        for box in results[0].boxes:
            conf = box.conf[0].item() 
            cls_id = box.cls[0].item()  
            if conf > confidence_threshold and cls_id == 0: 
                fire_detected = True

    return fire_detected

# Real-time fire detection with metrics collection
def real_time_fire_detection():
    global predictions

    cap = cv2.VideoCapture("C:\Users\rekha\OneDrive\Desktop\archive\Fire\Poly2\Poly2.MOV")  
    if not cap.isOpened():
        print("Error: Could not access the video.")
        return

    print("Starting fire detection...")

    frame_index = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Perform fire detection
        fire_detected = detect_fire_yolo(frame, yolo_model)
        predictions.append(int(fire_detected))  # Store prediction as binary (1 for fire, 0 for no fire)

        # Display results
        text = "Fire Detected!" if fire_detected else "No Fire"
        color = (0, 0, 255) if fire_detected else (0, 255, 0)
        cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
        cv2.imshow("Fire Detection", frame)

        # Quit if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        frame_index += 1
        if frame_index >= len(ground_truth):  # Stop when ground truth ends
            break

    cap.release()
    cv2.destroyAllWindows()

# Compute metrics and plot results
def compute_metrics_and_plot():
    global predictions

    # Ensure predictions align with ground truth length
    if len(predictions) != len(ground_truth):
        print("Error: Predictions and ground truth lengths do not match.")
        return

    # Compute metrics
    accuracy = accuracy_score(ground_truth, predictions)
    precision = precision_score(ground_truth, predictions)
    recall = recall_score(ground_truth, predictions)
    f1 = f1_score(ground_truth, predictions)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")

    # Plot metrics
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    values = [accuracy, precision, recall, f1]

    plt.figure(figsize=(10, 6))
    plt.bar(metrics, values, color=['blue', 'green', 'orange', 'red'])
    plt.ylim(0, 1)
    plt.title("Fire Detection Metrics")
    plt.ylabel("Score")
    for i, v in enumerate(values):
        plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontsize=12)
    plt.show()

if __name__ == "__main__":
    # Run fire detection
    real_time_fire_detection()

    # Compute and plot metrics
    compute_metrics_and_plot()
```
